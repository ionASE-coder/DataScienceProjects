Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

What are the dates of the earliest and most recent reviews? Store these values as two separate variables with your preferred names (review_date_early, review_date_recent)

```python
#importing tsv
df_l_review=pd.read_csv('data/airbnb_last_review.tsv', sep='\t')
print(df_l_review.info())

#last_review column is object. Let's transform it into date and time format
df_l_review['last_review']=pd.to_datetime(df_l_review['last_review'],format="%B %d %Y", errors='coerce')
print(df_l_review)
print(f'is na: {df_l_review.isna().sum()}')
#assert df_l_review['last_review'].dtype == 'datetime'
print(df_l_review.info())
#Saving the dates into a tuple
review_date_early=df_l_review['last_review'].min()
review_date_recent=df_l_review['last_review'].max()
print(review_date_early, review_date_recent)
```

Merging all three table on listing_id

```python
#importing airbnb_price.csv
df_price = pd.read_csv('data/airbnb_price.csv')
#importing excel file
xcel = pd.ExcelFile('data/airbnb_room_type.xlsx')
df_room = xcel.parse(xcel.sheet_names[0])
#Checking if its worth to join
print(f"airbnb_price has unique values: {df_price['listing_id'].unique()}, and a shape of {df_price.shape[0]}")
print(f"airbnb_room_type has unique values: {df_room['listing_id'].unique()}, and a shape of {df_room.shape[0]}")
print(f"airbnb_room_type has unique values: {df_l_review['listing_id'].unique()}, and a shape of {df_l_review.shape[0]}")

#merging all three
df=df_price.copy()
df = df.merge(df_room, on='listing_id')
df=df.merge(df_l_review, on='listing_id')
df
```

How many of the listings are private rooms? Save this into any variable.

```python
#sanity check for the room_type
print(df['room_type'].unique())
# there should be just 3 cats: entire home/apt, private room, shared room
#performing lower
df.info()
df['room_type']=df['room_type'].str.lower()
print(df['room_type'].unique())
#to categories
df['room_type']=pd.Categorical(df['room_type'])
private_rooms=df['room_type'].loc[df['room_type']=='private room'].count()
```

What is the average listing price? Round to the nearest two decimal places and save into a variable.

```python

#I want to see if there are dupicates ammong listing_id
duplicates=df.duplicated(subset=['listing_id'], keep=False)
duplicates=~duplicates
assert duplicates.any()==True
#converting price to float
print(df['price'].unique())
#dropping dollar
df['price']=df['price'].str.replace(" dollars","")
print(df['price'])
df['price']=df['price'].astype('float')
#mean
average_listing_price=df['price'].mean().round(2)
print(average_listing_price)
```

Combine the new variables into one DataFrame called review_dates with four columns in the following order: first_reviewed, last_reviewed, nb_private_rooms, and avg_price. The DataFrame should only contain one row of values.

```python
#putting all variables together in the desired format
#creating the dataframe
list=[[review_date_early, review_date_recent,private_rooms, average_listing_price]]
review_dates=pd.DataFrame(list, columns=['first_reviewed', 'last_reviewed','nb_private_rooms','avg_price'])
review_dates
```
